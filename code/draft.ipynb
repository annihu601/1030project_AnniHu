# Data set description
- target variable: weekly_sales




## features.csv
- contains additional data related to data, store, department, and regional activity for given data. 
- Store,Date,Temperature,Fuel_Price,MarkDown1,MarkDown2,MarkDown3,MarkDown4,MarkDown5,CPI,Unemployment,IsHoliday
- `Store`：Denotes the Store Number
- `Date`： Dates (Data) from 2010 - 2012
- `Temperature`：Temperature in Fahrenheit
- `Fuel_Price`： Price of Fuel that day
- `MarkDown1,MarkDown2,MarkDown3,MarkDown4,MarkDown5`：Anonymized data related to promotional markdowns at Walmart
- `CPI`：The consumer price index. It measures the monthly changes in prices paid by US consumers
- `Unemployment`： Rate of Unemployment
- `IsHoliday`： Is that particular day holiday or working day


## train.cvs
- This is the historical training data, which covers 2010–02–05 to 2012–11–01
- `Store`:The store number
- `Dept`: The department number
- `Date`:The week
- `Weekly_Sales`:Sales for the given department in the given store
- `IsHolidat`:Whether the week is a special holiday week

## test.csv
- This file is identical to train.csv, except  withheld the weekly sales.

## stores.csv
- this dataset contain anonymized information about 45 stores, indicating types and size of store 

- `Store`:Stores numbered from 1 to 45
- `Type`:Store type has been provided, there are 3 types — A, B and C
- `Size`: Stores size has provided
